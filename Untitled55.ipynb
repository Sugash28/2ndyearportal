````python
PAGE_SYSTEM_PROMPT = """Du är expert på svenska förfrågningsunderlag för kylanläggningar (AMA VVS & Kyl-format).
Extrahera strukturerad data från en sida i ett förfrågningsunderlag.
Returnera ENDAST giltig JSON — inga markdown-tecken, ingen inledning, ingen förklaring.

Om sidan ENDAST innehåller innehållsförteckning, signaturrader, administrativa
kontaktuppgifter eller blanka fält — returnera exakt: {"skip": true}

JSON-schema:
{
  "facility_type": sträng eller null,
  "refrigerants": [sträng],
  "control_system": sträng eller null,
  "num_cooling_rooms": heltal eller null,
  "num_freezer_rooms": heltal eller null,
  "num_display_units_cooling": heltal eller null,
  "num_display_units_freezer": heltal eller null,
  "display_unit_type": "öppen"|"stängd"|"blandat"|null,
  "room_temperatures": [
    {"room_id": sträng, "temp_c": heltal, "source_phrase": sträng}
  ],
  "heat_recovery": bool eller null,
  "comfort_cooling": bool eller null,
  "internet_gateway_required": bool eller null,
  "guarantee_years": heltal eller null,
  "special_conditions": [sträng],
  "requirements": [
    {
      "field_path": sträng,
      "value": sträng,
      "source_phrase": sträng,
      "confidence": "high"|"medium"|"low"
    }
  ]
}

field_path-konvention — använd EXAKT dessa prefix:
  kylrum.*        frysutrymme.*     kyldisk.*
  kylsystem.*     gaslarm.*         styrsystem.*
  certifiering.*  värmeåtervinning.*

requirements — ENDAST tekniska fakta för BOM:
✅ antal, temperaturer, fabrikat/modeller, larmtyper, certifieringar, köldmedier, tryck
❌ betalningsvillkor, garantiansvar, arbetstider, kontaktuppgifter

Regler:
- Extrahera ENDAST det som uttryckligen står på denna sida.
- Använd null / [] för fält som inte nämns.
- Hitta aldrig på eller gissa värden som inte finns i texten.
- source_phrase måste vara ett ordagrant citat från sidans text."""
````

````python
def is_empty_result(parsed):
    if parsed.get("skip"):
        return True
    top_fields = ["facility_type","refrigerants","control_system","num_cooling_rooms",
                  "num_freezer_rooms","num_display_units_cooling","num_display_units_freezer",
                  "heat_recovery","comfort_cooling","internet_gateway_required","guarantee_years"]
    return not (any(parsed.get(f) for f in top_fields)
                or parsed.get("special_conditions")
                or parsed.get("requirements")
                or parsed.get("room_temperatures"))

def run_page(page_text):
    prompt = f"[SYSTEM]\n{PAGE_SYSTEM_PROMPT}\n\n[USER]\n{page_text}"
    raw = model.invoke(prompt).strip()
    raw = re.sub(r"^```(?:json)?\s*", "", raw)
    raw = re.sub(r"\s*```$", "", raw)
    match = re.search(r"\{.*\}", raw, re.DOTALL)
    if match:
        raw = match.group(0)
    return json.loads(raw)

print("✅ Functions ready")
````

````python
results = {"project_name": data["project_name"], "documents": []}

for doc in data["documents"]:
    doc_result = {"source_file": doc["source_file"], "pages": {}}
    print(f"\n📄 {doc['source_file']}")

    for page_num, page_text in doc["pages"].items():
        if not page_text or len(page_text.strip()) < 50:
            print(f"  Page {page_num} ⏭ skipped (empty)")
            continue
        print(f"  Page {page_num}...", end=" ", flush=True)
        try:
            parsed = run_page(page_text)
            if is_empty_result(parsed):
                print("⏭ no relevant data")
            else:
                doc_result["pages"][page_num] = parsed
                print(f"✅ {len(parsed.get('requirements', []))} requirements")
        except Exception as e:
            print(f"❌ {e}")

    results["documents"].append(doc_result)

print("\n✅ Processing complete!")
````

````python
def build_full_bom_context(df_bom):
    """Give LLM the complete BOM — no filtering, no hints needed."""
    lines = []
    for _, row in df_bom.iterrows():
        code  = str(row.get("article_code") or "")
        name  = str(row.get("name")  or "")
        specs = str(row.get("specs") or "")
        price = str(row.get("total_price") or "")
        if code and name and code != "nan":
            lines.append(f"- {code} | {name} | {specs} | Price: {price}")
    return "\n".join(lines)

BOM_CONTEXT = build_full_bom_context(df_bom)
print("✅ BOM context built:")
print(BOM_CONTEXT)
````

````python
from langchain_ollama.llms import OllamaLLM
import json, re

model = OllamaLLM(model="llama3")

def llm_map_requirement(category, field_path, value, source_phrase):
    """
    Let the LLM decide the mapping completely.
    No hardcoded keywords or article codes needed.
    """
    prompt = f"""You are an expert in refrigeration and cooling installations (Swedish: kylanläggningar).

Your job is to match an RFQ requirement to the most relevant BOM article.

RFQ Requirement:
  Category:     {category}
  Field path:   {field_path}
  Value:        {value}
  Source text:  {source_phrase}

Available BOM articles (code | name | specs | price):
{BOM_CONTEXT}

Instructions:
- Read the requirement carefully including the source text
- Find the most relevant BOM article that would fulfill this requirement
- The match can be indirect (e.g. a temperature requirement → temperature sensor)
- If multiple articles match, pick the most specific one
- If nothing matches, return null

Return ONLY this JSON, no explanation:
{{
  "article_code": string or null,
  "article_name": string or null,
  "confidence": "high"|"medium"|"low",
  "indirect_match": true|false,
  "reason": string,
  "quantity_hint": number or null
}}"""

    try:
        raw = model.invoke(prompt).strip()
        raw = re.sub(r"^```(?:json)?\s*", "", raw)
        raw = re.sub(r"\s*```$", "", raw)
        match = re.search(r"\{.*\}", raw, re.DOTALL)
        if match:
            raw = match.group(0)
        return json.loads(raw)
    except Exception as e:
        return {
            "article_code": None, "article_name": None,
            "confidence": "low", "indirect_match": False,
            "reason": f"Error: {e}", "quantity_hint": None
        }

print("✅ LLM mapper ready")
````

````python
def repair_json(text):
    # Remove markdown code block fences if present at start/end
    text = re.sub(r"^```(?:json)?\s*", "", text)
    text = re.sub(r"\s*```$", "", text)

    try:
        stack = []
        json_start = -1
        json_end = -1

        for i, char in enumerate(text):
            if char == '{':
                if json_start == -1:
                    json_start = i
                stack.append('{')
            elif char == '}':
                if stack:
                    stack.pop()
                    if not stack and json_start != -1: # Found a balanced pair
                        json_end = i
                        break
                elif json_start == -1:
                    continue # Ignore '}' if no '{' has been opened

        if json_start != -1 and json_end != -1:
            extracted_json_str = text[json_start : json_end + 1]
            try:
                json.loads(extracted_json_str)
                return extracted_json_str
            except json.JSONDecodeError:
                # Attempt to remove trailing commas within arrays and objects
                cleaned_json_str = re.sub(r",\s*([\]}])", r"\1", extracted_json_str)
                json.loads(cleaned_json_str) # This will re-raise if still invalid
                return cleaned_json_str
    except Exception:
        pass
    return "{}"

print("✅ repair_json function defined")
````

````python
import json
import re
import requests

# ── Merge requirements into component projects ────────────────────────────────

def inject_requirements(components, req_profiles):
    enriched = []
    for comp in components:
        new_comp = dict(comp)
        new_projects = {}
        for proj_name, proj_data in comp.get("projects", {}).items():
            entry = dict(proj_data)
            entry["requirements"] = (
                req_profiles.get(proj_name, {}).get("requirements", [])
            )
            new_projects[proj_name] = entry
        new_comp["projects"] = new_projects
        enriched.append(new_comp)
    return enriched


# ── Build focused context per component ──────────────────────────────────────

def build_context(component):
    historical_evidence  = {}
    project_requirements = {}
    for proj_name, proj_data in component.get("projects", {}).items():
        historical_evidence[proj_name] = {
            "quantity":    proj_data.get("quantity"),
            "description": proj_data.get("description"),
            "category":    proj_data.get("category"),
        }
        reqs = proj_data.get("requirements", [])
        if reqs:
            project_requirements[proj_name] = reqs
    return {
        "target_component": {
            "article_code": component.get("article_code"),
            "description":  component.get("description"),
            "r_squared":    component.get("r_squared"),
            "top_features": component.get("top_features", []),
        },
        "historical_evidence":  historical_evidence,
        "project_requirements": project_requirements,
    }


# ── Ollama ────────────────────────────────────────────────────────────────────

RULE_SYSTEM_PROMPT = """Du ar expert pa deterministiska regler for kylanlaggningar.
Skapa en specifik regel for komponenten baserat pa dess historiska anvandning och projektkrav.
Returnera ENDAST giltig JSON:
{
  "rule": {
    "article_code": "...",
    "description": "...",
    "condition": {"feature": "...", "operator": "eq|gt|lt|contains", "value": "..."},
    "quantity_formula": "...",
    "confidence": "high|medium|low",
    "rationale": "..."
  }
}"""


def call_ollama(user_content, model="llama3"):
    try:
        resp = requests.post(
            "http://localhost:11434/api/chat",
            json={
                "model": model,
                "stream": False,
                "options": {"num_predict": 4096, "temperature": 0},
                "messages": [
                    {"role": "system", "content": RULE_SYSTEM_PROMPT},
                    {"role": "user",   "content": user_content},
                ],
            },
            timeout=120,
        )
        resp.raise_for_status()
        return resp.json()["message"]["content"]
    except Exception as e:
        return f"ERROR: {e}"


def parse_rule(raw):
    try:
        match = re.search(r"\{.*\}", raw, re.DOTALL)
        if match:
            payload = json.loads(match.group(0))
            return payload.get("rule", payload)
    except json.JSONDecodeError:
        pass
    return None


# ── Run ───────────────────────────────────────────────────────────────────────
# scaling_components and profiles come from earlier cells in your notebook

scaling_enriched = inject_requirements(scaling_components, profiles)

all_rules = []
total = len(scaling_enriched)

for i, component in enumerate(scaling_enriched, 1):
    code = component.get("article_code", "?")
    print(f"[{i:>3}/{total}] {code}", end="  ", flush=True)

    context  = build_context(component)
    raw_resp = call_ollama(json.dumps(context, ensure_ascii=False, indent=2))

    if raw_resp.startswith("ERROR"):
        print(f"ERROR: {raw_resp}")
        continue

    rule = parse_rule(raw_resp)
    if rule:
        rule.setdefault("article_code", code)
        all_rules.append(rule)
        print("OK")
    else:
        print("parse failed")

with open("refined_rules.json", "w", encoding="utf-8") as f:
    json.dump({"rules": all_rules}, f, indent=2, ensure_ascii=False)

print(f"\nDone! {len(all_rules)}/{total} rules saved to refined_rules.json")
````

````python
import json
import numpy as np

# ── FUNCTION 1: inject_requirements ──────────────────────────────────────────

def inject_requirements(components, req_profiles):
    enriched = []
    for comp in components:
        new_comp = dict(comp)
        new_projects = {}
        # quantities is {proj_name: qty_value} — convert to projects dict
        for proj_name, qty in comp.get("quantities", {}).items():
            new_projects[proj_name] = {
                "project_name": proj_name,
                "quantity":     qty,
                "requirements": req_profiles.get(proj_name, {}).get("requirements", [])
            }
        new_comp["projects"] = new_projects
        enriched.append(new_comp)
    return enriched


# ── FUNCTION 2: extract_features ─────────────────────────────────────────────

def extract_features(requirements):
    feats = {
        "refrigerant_co2":   0,
        "refrigerant_hfc":   0,
        "refrigerant_r290":  0,
        "num_cooling_rooms": 0,
        "num_freezer_rooms": 0,
        "facility_retail":   0,
        "facility_school":   0,
        "facility_care":     0,
    }
    for req in requirements:
        fp  = req.get("field_path", "").lower()
        val = str(req.get("value",  "")).lower()
        if any(x in val for x in ["co2","r744","koldioxid"]): feats["refrigerant_co2"] = 1
        if any(x in val for x in ["r290","propan"]):          feats["refrigerant_r290"] = 1
        if any(x in val for x in ["r448","r513","hfc"]):      feats["refrigerant_hfc"] = 1
        if "kylrum.antal" in fp:
            try:    feats["num_cooling_rooms"] = max(feats["num_cooling_rooms"], int(float(val)))
            except: pass
        if "frysrum.antal" in fp:
            try:    feats["num_freezer_rooms"] = max(feats["num_freezer_rooms"], int(float(val)))
            except: pass
        if any(x in fp+val for x in ["butik","retail","ica","hemköp","willys"]): feats["facility_retail"] = 1
        if any(x in fp+val for x in ["skola","förskola","school"]):              feats["facility_school"] = 1
        if any(x in fp+val for x in ["vård","omsorg","solgården","care"]):       feats["facility_care"]   = 1
    return feats


# ── FUNCTION 3: fit_rule_dynamic ──────────────────────────────────────────────

def fit_rule_dynamic(component, previous_rule=None):
    code      = component["article_code"]
    top_feats = component.get("top_features", [])

    # read from projects dict built by inject_requirements
    data_points = []
    for proj_name, proj_data in component.get("projects", {}).items():
        qty = proj_data.get("quantity")
        if not qty or qty <= 0:
            continue
        reqs  = proj_data.get("requirements", [])
        feats = extract_features(reqs)
        data_points.append({"project": proj_name, "qty": float(qty), "features": feats})

    n = len(data_points)
    print(f"       → {n} projects with qty>0", end="  ")

    if n < 3:
        fallback = round(float(np.mean([d["qty"] for d in data_points])), 1) if n > 0 else 0
        return {
            "article_code":      code,
            "description":       component.get("description", ""),
            "quantity_formula":  str(int(round(fallback))),
            "confidence":        "none",
            "confidence_reason": f"only {n} project(s) with qty>0 — need at least 3",
            "n_projects":         n,
            "mae":               None,
            "quantity_stats":    {"mean": fallback, "min": 0, "max": 0},
        }

    y = np.array([d["qty"] for d in data_points], dtype=float)

    # drop zero-variance features
    feature_names = []
    for feat in top_feats[:2]:
        f    = feat["feature"]
        vals = [d["features"].get(f, 0) for d in data_points]
        if len(set(vals)) > 1:
            feature_names.append(f)

    # all features constant → use mean
    if not feature_names:
        mean_qty = round(float(np.mean(y)), 1)
        mae      = round(float(np.mean(np.abs(y - mean_qty))), 2)
        return {
            "article_code":      code,
            "description":       component.get("description", ""),
            "n_projects":        n,
            "features_used":     [],
            "base":              mean_qty,
            "coefficients":      {},
            "quantity_formula":  str(int(round(mean_qty))),
            "quantity_stats":    {"mean": mean_qty, "min": int(min(y)), "max": int(max(y))},
            "mae":               mae,
            "coeff_drift":       None,
            "confidence":        "low",
            "confidence_reason": "all features constant — using mean qty",
        }

    # least squares
    X     = np.array([[d["features"].get(f, 0) for f in feature_names]
                       for d in data_points], dtype=float)
    X_int = np.column_stack([np.ones(n), X])
    coeffs, _, _, _ = np.linalg.lstsq(X_int, y, rcond=None)
    base    = round(float(coeffs[0]), 2)
    co_arr  = [round(float(c), 3) for c in coeffs[1:]]
    min_qty = int(min(y))

    terms   = " + ".join(f"{c}*{f}" for c, f in zip(co_arr, feature_names))
    formula = f"max({min_qty}, round({base} + {terms}))"

    errors = [abs(max(min_qty, round(base + sum(c*d["features"].get(f,0)
              for c,f in zip(co_arr,feature_names)))) - d["qty"])
              for d in data_points]
    mae = round(sum(errors)/n, 2)

    drift = None
    if previous_rule and "coefficients" in previous_rule:
        prev  = previous_rule["coefficients"]
        drift = round(sum(abs(new - prev.get(f,0))
                          for new,f in zip(co_arr,feature_names)), 3)

    if   mae > 10:  confidence, reason = "none",   f"MAE={mae} — unreliable, review manually"
    elif mae > 3:   confidence, reason = "low",    f"MAE={mae} — off by ~{mae:.0f} units avg"
    elif mae > 1.5: confidence, reason = "medium", f"MAE={mae} — usable, add more projects"
    elif drift is not None and drift <= 0.1:
                    confidence, reason = "high",   f"MAE={mae}, drift={drift} — converged"
    else:           confidence, reason = "high",   f"MAE={mae} — accurate"

    return {
        "article_code":      code,
        "description":       component.get("description", ""),
        "n_projects":        n,
        "features_used":     feature_names,
        "base":              base,
        "coefficients":      dict(zip(feature_names, co_arr)),
        "quantity_formula":  formula,
        "quantity_stats":    {"mean": round(float(np.mean(y)),1),
                              "min":  min_qty, "max": int(max(y))},
        "mae":               mae,
        "coeff_drift":       drift,
        "confidence":        confidence,
        "confidence_reason": reason,
    }


# ── RUN ───────────────────────────────────────────────────────────────────────

scaling_enriched = inject_requirements(scaling_components, profiles)

all_rules      = []
previous_rules = {}
total          = len(scaling_enriched)

for i, component in enumerate(scaling_enriched, 1):
    code = component.get("article_code", "?")
    print(f"[{i:>3}/{total}] {code}", end="  ", flush=True)

    rule = fit_rule_dynamic(component, previous_rule=previous_rules.get(code))

    all_rules.append(rule)
    previous_rules[code] = rule
    print(f"[{rule['confidence'].upper()}]  MAE={rule['mae']}  {rule['quantity_formula']}")

# ── SAVE ──────────────────────────────────────────────────────────────────────

with open("refined_rules.json", "w", encoding="utf-8") as f:
    json.dump({"rules": all_rules}, f, indent=2, ensure_ascii=False)

print(f"\nDone! {len(all_rules)}/{total} rules saved to refined_rules.json")

high   = [r for r in all_rules if r["confidence"] == "high"]
medium = [r for r in all_rules if r["confidence"] == "medium"]
low    = [r for r in all_rules if r["confidence"] == "low"]
none_  = [r for r in all_rules if r["confidence"] == "none"]

print(f"\nConfidence breakdown:")
print(f"  HIGH   {len(high):>3}  use directly for BOM")
print(f"  MEDIUM {len(medium):>3}  verify with engineer")
print(f"  LOW    {len(low):>3}  add more projects")
print(f"  NONE   {len(none_):>3}  manual review needed")

if none_:
    print(f"\nNeeds manual review:")
    for r in none_:
        print(f"  {r['article_code']:<25}  MAE={r['mae']}  n={r['n_projects']}")
````

````python
import numpy as np

def extract_features(requirements):
    feats = {
        "refrigerant_co2":   0, "refrigerant_hfc":   0, "refrigerant_r290":  0,
        "num_cooling_rooms": 0, "num_freezer_rooms": 0,
        "facility_retail":   0, "facility_school":   0, "facility_care":     0,
    }
    for req in requirements:
        fp  = req.get("field_path", "").lower()
        val = str(req.get("value",  "")).lower()
        if any(x in val for x in ["co2","r744","koldioxid"]): feats["refrigerant_co2"] = 1
        if any(x in val for x in ["r290","propan"]):          feats["refrigerant_r290"] = 1
        if any(x in val for x in ["r448","r513","hfc"]):      feats["refrigerant_hfc"] = 1
        if "kylrum.antal" in fp:
            try:    feats["num_cooling_rooms"] = max(feats["num_cooling_rooms"], int(float(val)))
            except: pass
        if "frysrum.antal" in fp:
            try:    feats["num_freezer_rooms"] = max(feats["num_freezer_rooms"], int(float(val)))
            except: pass
        if any(x in fp+val for x in ["butik","retail","ica","hemköp","willys"]): feats["facility_retail"] = 1
        if any(x in fp+val for x in ["skola","förskola","school"]):              feats["facility_school"] = 1
        if any(x in fp+val for x in ["vård","omsorg","äldreboende","care"]):     feats["facility_care"]   = 1
    return feats


def predict_and_show(test_project, all_rules, base_components, profiles, bom_data):

    reqs  = profiles.get(test_project, {}).get("requirements", [])
    feats = extract_features(reqs)

    # sum actual quantities for comparison
    actual_qty = {}
    for item in bom_data.get(test_project, []):
        c = item["article_code"]
        actual_qty[c] = actual_qty.get(c, 0) + item["quantity"]

    # ── predict ───────────────────────────────────────────────────────────────
    results = []

    for base in base_components:
        code = base["article_code"]
        results.append({
            "article_code":  code,
            "description":   base.get("note", ""),
            "predicted_qty": 1,
            "actual_qty":    actual_qty.get(code),
            "confidence":    "BASE",
            "mae":           None,
        })

    for rule in all_rules:
        code      = rule["article_code"]
        base_val  = rule.get("base", 0)
        coeffs    = rule.get("coefficients", {})
        min_qty   = rule.get("quantity_stats", {}).get("min", 0)
        raw       = base_val + sum(c * feats.get(f, 0) for f, c in coeffs.items())
        predicted = max(min_qty, round(raw))
        results.append({
            "article_code":  code,
            "description":   rule.get("description", ""),
            "predicted_qty": predicted,
            "actual_qty":    actual_qty.get(code),
            "confidence":    rule["confidence"].upper(),
            "mae":           rule.get("mae"),
        })

    # sort by confidence then article code
    order = {"BASE":0, "HIGH":1, "MEDIUM":2, "LOW":3, "NONE":4}
    results.sort(key=lambda x: (order.get(x["confidence"], 5), x["article_code"]))

    # ── ONLY show components that exist in actual BOM ─────────────────────────
    # filter: keep only rows where actual_qty is known (component used in this project)
    relevant = [r for r in results if r["actual_qty"] is not None]

    # ── print ─────────────────────────────────────────────────────────────────
    print(f"\n{'═'*65}")
    print(f"  PREDICTED BOM — {test_project}")
    print(f"{'═'*65}")
    print(f"  Features: " + "  ".join(f"{k}={v}" for k,v in feats.items() if v))
    print()
    print(f"  {'ARTICLE CODE':<25} {'DESCRIPTION':<28} {'QTY':>4}  CONFIDENCE")
    print(f"  {'─'*62}")

    conf_labels = {
        "BASE":   "── base components",
        "HIGH":   "── high confidence",
        "MEDIUM": "── medium confidence",
        "LOW":    "── low confidence",
        "NONE":   "── needs review",
    }

    current_conf = None
    exact = 0
    in_mae = 0
    total_qty = 0

    for r in relevant:
        # print section header when confidence changes
        if r["confidence"] != current_conf:
            current_conf = r["confidence"]
            print(f"\n  {conf_labels[current_conf]}")

        qty    = r["predicted_qty"]
        desc   = r["description"][:26]
        actual = r["actual_qty"]
        mae    = r["mae"]

        # accuracy icon
        if qty == int(actual):
            icon = "✓"
            exact += 1
        elif mae and abs(qty - actual) <= mae:
            icon = "~"
            in_mae += 1
        else:
            icon = "!"

        print(f"  {r['article_code']:<25} {desc:<28} {qty:>4}  {r['confidence']}  {icon}")
        total_qty += qty

    # ── summary ───────────────────────────────────────────────────────────────
    n = len(relevant)
    accuracy = round((exact + in_mae) / n * 100) if n else 0

    print(f"\n  {'─'*62}")
    print(f"  {n} components   {total_qty} units total")
    print(f"  ✓ {exact} exact   ~ {in_mae} within MAE   ! {n-exact-in_mae} outside   →  {accuracy}% accuracy")
    print(f"{'═'*65}\n")

    return relevant


# ── RUN ───────────────────────────────────────────────────────────────────────
# change TEST_PROJECT to any project name

TEST_PROJECT = "Tellusgatan"
# options: Amhults_FSK, Hemköp_Malmö, ICA_Kvantum__Flen, ICA_Kvantum__Nybro,
#          Ingaredsskolan, Jonslunds_Skola, Snöbäret, Solgården,
#          Tellusgatan, Västhojraskolan

predict_and_show(
    test_project    = TEST_PROJECT,
    all_rules       = all_rules,
    base_components = base_components,
    profiles        = profiles,
    bom_data        = bom_data,
)
````
